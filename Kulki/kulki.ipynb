```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import math
import os
import sys

path_utils = os.path.join(r"C:\Users\qmiko\Desktop\python_projekty\Kulki\utils")
sys.path.append(os.path.abspath(path_utils))

from animation import update, init
from kulka import Kulka, check_colission, check_wall, main
```

```python
dt = 0.1
a = 25

fig, ax = plt.subplots()
ax.set_xlim(0, a)
ax.set_ylim(0, a)
ax.set_aspect("equal")
ax.set_xticks([])
ax.set_yticks([])
ax.set_title("Zderzenia sprężyste")

global kulki
kulki, T = main(a, dt)
circles = init(ax, kulki)

ani = animation.FuncAnimation(
    fig,
    update,
    frames=int((T*5)/(dt)),
    init_func= lambda: init(ax, kulki),
    fargs=(circles, kulki, dt, a),
    interval=20,
    blit=False,
    repeat=True
)
ani.save('kulki_duzo2.gif', writer = 'pillow', fps = 30)
plt.show()
```
